```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("airlines_flights_data.csv")

# Exploratory Data Analysis (EDA) - Pie Chart Most Context for Nonconditional Data
target_col = ['source_city','departure_time','stops','arrival_time','destination_city','class']

for dt in target_col:
    # Dataset
    top_context = df[dt].value_counts()
    
    # Pie Chart
    plt.figure(figsize=(8,8))
    top_context.plot.pie(autopct='%1.1f%%', startangle=90)
    plt.title(f"{dt.replace("_"," ").title()} Comparison")
    plt.ylabel("")
    plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Comparison Of Flight Duration Category
df = pd.read_csv("airlines_flights_data.csv")

# Convert duration
def categorize_duration(time):
    if time <= 3:
        return "Short Flight"
    elif time <= 7:
        return "Medium Flight"
    elif time <= 12:
        return "Long Flight"
    elif time <= 24:
        return "A Day"
    else:
        return "More than A Day"

df["Flight Duration Category"] = df["duration"].apply(categorize_duration)

# Count the number of flight
quartal_counts = df["Flight Duration Category"].value_counts().sort_index()

# Pie Chart
plt.figure(figsize=(6, 6))
quartal_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Comparison Of Flight Duration Category")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Comparison Of Flight Price Category
df = pd.read_csv("airlines_flights_data.csv")

# Convert Price
def categorize_price(price):
    if price <= 5000:
        return "Cheap"
    elif price <= 10000:
        return "Moderate"
    elif price <= 25000:
        return "Expensive"
    else:
        return "Very Expensive"

df["Flight Price Category"] = df["price"].apply(categorize_price)

# Count the number of flight
quartal_counts = df["Flight Price Category"].value_counts().sort_index()

# Pie Chart
plt.figure(figsize=(6, 6))
quartal_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Comparison Of Flight Price Category")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Comparison Of Arrival Time (AM vs PM)
df = pd.read_csv("airlines_flights_data.csv")

# Convert Price
def categorize_arrival(arrival):
    if arrival in ("Night", "Afternoon", "Evening", "Late_Night"):
        return "PM"
    else:
        return "AM"

df["Arrival Time Category"] = df["arrival_time"].apply(categorize_arrival)

# Count the number of flight
quartal_counts = df["Arrival Time Category"].value_counts().sort_index()

# Pie Chart
plt.figure(figsize=(6, 6))
quartal_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Comparison Of Arrival Time (AM vs PM)")
plt.ylabel("")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
# Exploratory Data Analysis (EDA) - Pie Chart Comparison Of Flight Type
df = pd.read_csv("airlines_flights_data.csv")

# Convert Price
def categorize_flight(flight):
    if "-" in flight:
        parts = flight.split("-")
        len_code = len(parts[1])
        
        if len_code == 3:
            return "Regular"
        elif len_code == 4:
            return "Charter / Ferry"

df["Flight Type"] = df["flight"].apply(categorize_flight)

# Count the number of flight
quartal_counts = df["Flight Type"].value_counts().sort_index()

# Pie Chart
plt.figure(figsize=(6, 6))
quartal_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.title("Comparison Of Flight Type")
plt.ylabel("")
plt.tight_layout()
plt.show()
```
